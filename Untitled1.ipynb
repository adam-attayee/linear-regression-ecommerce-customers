## Splitting the Data into independent (X) and dependent (y) variables

```python
df.columns
```

```python
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

## Predicting the Test Data

```python
prediction = lm.predict(X_test)
```

```python
#scatterplot of the real test values versus the predicted values.
plt.grid()
plt.scatter(y_test, prediction)
plt.xlabel("Y Test")
plt.ylabel("Prediction")
```

## Evaluating the Model

```python
from sklearn import metrics 
```

```python
mae = metrics.mean_absolute_error(y_test, prediction)
mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
#The higher this value is, the better the model fits the data. A value of .989 means that 99% of the variation in the price is
#explained by the model.
r2 = metrics.explained_variance_score(y_test, prediction)
print(f"R-squared: {r2}")
```

## Residuals

```python
#The residuals are normally distributed, this is an indication that the model is a good fit.
sns.displot(y_test - prediction, bins=50, kde=True)
plt.grid()
plt.ylabel(None)
plt.show()
```

## Mobile app or website?

```python
coef_df = pd.DataFrame(lm.coef_, X.columns, columns=["Coefficients"])
coef_df
```

The coefficients of the model tells us how much we expect total spending to increase by given a one-unit increase in the variables, all other things being equal. 

When time spent on the app goes up by one unit, we expect total spending to increase by approximately \\$39, and when time on website goes up by one unit, we expect total spending to increase by \\$0.19. 

It might be tempting to say that the company should focus on their mobile app since prolonging a customer's time on the app seem to increase spending a lot more than the website, but it is not possible to make that decision with the data provided and without more knowledge about the operations of the business, i.e. the costs associated with developing the app vs the website. 

I will quickly analyze the relationship between length of membership and whether long-term members prefer to use the app or the website since length of membership seem to be the greatest driver of spending

## Do long-term members prefer the app or the website?
